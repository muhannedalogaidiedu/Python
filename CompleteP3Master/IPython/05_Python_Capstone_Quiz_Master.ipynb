## Final Capstone Project

### Mini Python Utility Suite
Build a small package with:
1) **Regex Email Extractor** — module `extract.py` with `extract_emails(text)->list[str]`  
2) **File Reader with Exception Handling** — module `files.py` with `read_text(path)->str` that returns `""` on `FileNotFoundError`  
3) **OOP Manager** — class `Suite` in `suite.py` that wires the two, with a CLI-like `run(path)` method

**Bonus**: Add unit tests with `unittest` or `pytest`.

```
# Starter stubs
import re, os

def extract_emails(text:str):
    return re.findall(r"[\w.-]+@[\w.-]+\.[A-Za-z]{2,}", text or "")

def read_text(path:str):
    try:
        with open(path, "r", encoding="utf-8") as f:
            return f.read()
    except FileNotFoundError:
        return ""

class Suite:
    def __init__(self): pass
    def run(self, path):
        data = read_text(path)
        emails = extract_emails(data)
        return {"path": path, "count": len(emails), "emails": emails}

print(Suite().run("nonexistent.txt"))  # demo
```
