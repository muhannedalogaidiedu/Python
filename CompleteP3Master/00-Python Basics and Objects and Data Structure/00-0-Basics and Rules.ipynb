### Case sensitivity
Syntax refers to the set of rules that defines the combinations of symbols that are considered to be correctly structured programs in a language. In simpler terms, syntax is about the correct arrangement of words and symbols in a code.

Semantics refers to the meaning or the interpretation of the symbols, characters, and commands in a language. It is about what the code is supposed to do when it runs.

```python
## Basic Syntax Rules In Python
## Case sensitivity- Python is case sensitive

name="Doe"
Name="Joe"

print(name)
print(Name)
```

### Escape Characters

In Python, escape characters are special sequences used within strings to represent characters that cannot be directly typed or have a special meaning within the string syntax. These sequences begin with a backslash (`\`) followed by one or more characters. [1]  
Here are some common Python escape characters and their meanings: 

• \': Represents a single quote, allowing you to include a single quote within a single-quoted string. 

    print('It\'s a beautiful day.')

• \": Represents a double quote, allowing you to include a double quote within a double-quoted string. 

    print("He said, \"Hello!\"")

• \\: Represents a literal backslash character. 

    print("This is a backslash: \\")

• \n: Represents a newline character, moving the cursor to the beginning of the next line. [2]  

    print("Line 1\nLine 2")

• \t: Represents a horizontal tab character, inserting a tab space. 

    print("Name:\tJohn Doe")

• \r: Represents a carriage return, moving the cursor to the beginning of the current line. [2]  

    print("Text 1\rText 2") # This will likely output "Text 2" as "Text 1" is overwritten

• \b: Represents a backspace character. 

    print("Hello\bWorld") # This will likely output "HellWorld"

• \f: Represents a form feed character. 
• \ooo: Represents a character with an octal value (e.g., \101 for 'A'). 
• \xhh: Represents a character with a hexadecimal value (e.g., \x41 for 'A'). 

### Raw Strings: 

To prevent Python from interpreting escape sequences, you can use raw strings by prefixing the string literal with an r or R. In raw strings, backslashes are treated as literal characters, except for a backslash at the end of the string if it's not escaped itself. 

```python
# Normal string with escape sequence
normal_string = "C:\\Users\\Documents"
print(normal_string)

# Raw string, backslashes are literal
raw_string = r"C:\Users\Documents"
print(raw_string)
```
