9. How to exit for loop operations?
Tags - #forloop #exit #looping #concept

Break statement: It will terminate the loop and transfer the execution to the statement. 

Example 

#For loop with break

for Val in "GreatLearning":
  if Val == "L":
     break
  print(Val)
print('The End')
Output: 

G
r
e
a
t
The end
Here, you can see that the for loop will iterate over the elements in the input string until it encounters ‘L’. The break statement will play a role here and terminate the loop with a print statement.

Continue statement: It will allow you to skip the remainder and reset the condition.

#For loop with continue 

for Val in "GreatLearning":
   if Val == "L":
      continue
   print(Val)
print('The End')

Output: G r e a t e a r n i n g
In this output, you can see that the continue statement skips the specified iteration and executes the rest of the iterations.
